```python
# ============================================================
# IPL BATTER VALUATION MODEL
# ============================================================

# BUSINESS PROBLEM:
#
# IPL franchises invest crores of rupees during player auctions.
# However, selecting the right batter at the right price is a
# major challenge. Overpaying for an underperforming player or
# underestimating a consistent performer can negatively impact
# team performance and budget allocation.
#
# To solve this problem, a Data-Driven Batter Valuation Model
# was developed using historical IPL performance data.

# ============================================================
# OBJECTIVE
# ============================================================
#
# To estimate the fair auction value of a batter based on:
# 1. Total Runs Scored
# 2. Strike Rate
# 3. Matches Played (Experience)
#
# The model helps franchises make informed auction decisions
# instead of relying only on reputation or intuition.

# ============================================================
# FEATURES USED
# ============================================================
#
# 1. TOTAL RUNS (Weight: 60%)
#    - Measures consistency and batting performance.
#    - Higher runs indicate greater contribution to the team.
#
# 2. STRIKE RATE (Weight: 25%)
#    - Measures scoring speed in T20 cricket.
#    - Higher strike rate indicates aggressive and impactful batting.
#
# 3. MATCHES PLAYED (Weight: 15%)
#    - Represents player experience.
#    - Experienced players are generally more reliable in
#      high-pressure situations.

# ============================================================
# PLAYER SCORE CALCULATION
# ============================================================
#
# All performance metrics were normalized using MinMax Scaling
# to bring them onto the same scale.
#
# Batter Score Formula:
#
# Batter Score =
# (Runs Score × 60%)
# + (Strike Rate Score × 25%)
# + (Match Experience Score × 15%)
#
# Final score is calculated on a scale of 0–100.

# ============================================================
# AUCTION PRICE ESTIMATION
# ============================================================
#
# IPL auctions have a minimum base price for players.
#
# Assumptions:
# Minimum Auction Price = ₹20 Lakh (0.20 Cr)
# Maximum Auction Price = ₹25 Crore
#
# Estimated Auction Price Formula:
#
# Estimated Price (Cr) =
# 0.20 + (Batter Score / 100) × 24.80
#
# This ensures:
# - No player value falls below the IPL base price.
# - Elite players can reach realistic IPL valuation ranges.

# ============================================================
# BUSINESS INSIGHTS
# ============================================================
#
# Players with:
# - High run accumulation
# - Strong strike rates
# - Significant IPL experience
#
# receive higher valuation scores and estimated auction prices.
#
# Example:
# Virat Kohli achieved the highest valuation due to:
# - Highest run tally
# - Consistent performance
# - Extensive IPL experience
#
# Estimated Auction Value:
# ₹21+ Crore

# ============================================================
# CONCLUSION
# ============================================================
#
# This Batter Valuation Model converts historical player
# performance into an estimated auction value.
#
# The model provides franchises with a data-driven framework
# for evaluating batters and reducing investment risk during
# IPL auctions.
#
# Instead of relying solely on popularity or subjective
# judgment, teams can use objective performance metrics to
# make smarter auction decisions.
#
# ============================================================
```

```python
# IPL BOWLER VALUATION MODEL
```

```python
# BOWLER DATASET

# Ensure numeric columns

df['runs_bowler'] = pd.to_numeric(df['runs_bowler'], errors='coerce')
df['valid_ball'] = pd.to_numeric(df['valid_ball'], errors='coerce')
df['bowler_wicket'] = pd.to_numeric(df['bowler_wicket'], errors='coerce')

# Bowler Statistics

bowler_stats = df.groupby('bowler').agg(
    wickets=('bowler_wicket', 'sum'),
    runs_conceded=('runs_bowler', 'sum'),
    balls_bowled=('valid_ball', 'sum'),
    matches=('match_id', 'nunique')
).reset_index()

bowler_stats.head()
```

```python
# ECONOMY RATE
# Economy Rate = Total Runs Conceded / Overs Bowled

bowler_stats['overs_bowled'] = (
    bowler_stats['balls_bowled'] / 6
)

bowler_stats['economy'] = (
    bowler_stats['runs_conceded'] /
    bowler_stats['overs_bowled']
)

bowler_stats.head()
```

```python
# ============================================================
# ECONOMY RATE CALCULATION
# ============================================================

# Economy Rate is one of the most important bowling metrics
# in T20 cricket. It measures how many runs a bowler concedes
# on average per over.

# Formula:
# Economy Rate = Runs Conceded / Overs Bowled

# Step 1:
# Convert total balls bowled into overs.
# Since 1 over consists of 6 balls:
#
# Overs Bowled = Balls Bowled / 6

# Step 2:
# Calculate economy rate using:
#
# Economy = Runs Conceded / Overs Bowled

# Example:
# Runs Conceded = 240
# Balls Bowled = 180
#
# Overs Bowled = 180 / 6 = 30
# Economy = 240 / 30 = 8.0

# Interpretation:
# Economy 6-7   = Excellent
# Economy 7-8   = Good
# Economy 8-9   = Average
# Economy > 9   = Poor

# Lower economy indicates better bowling performance
# because the bowler is restricting runs effectively.

# Since lower economy is better, reverse scoring was
# applied while calculating the final bowler score so
# that bowlers with lower economy receive higher scores.
```

```python
#Genuine Bowlers
bowlers = bowler_stats[
    bowler_stats['wickets'] >= 20
].copy()

bowlers.shape
```

```python
#Check Top Bowlers
bowlers.sort_values(
    'wickets',
    ascending=False
)[[
    'bowler',
    'wickets',
    'economy',
    'matches'
]].head(15)
```

```python
#Normalize Features
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

bowlers[['wicket_score',
         'match_score']] = scaler.fit_transform(
    bowlers[['wickets',
             'matches']]
)
```

```python
#Economy Score (Reverse)
bowlers['economy_score'] = (
    bowlers['economy'].max() -
    bowlers['economy']
)

bowlers[['economy_score']] = scaler.fit_transform(
    bowlers[['economy_score']]
)
```

```python
#Bowler Score
# Wickets = 50%
# Economy = 30%
# Matches = 20%

bowlers['bowler_score'] = (
    bowlers['wicket_score'] * 0.50 +
    bowlers['economy_score'] * 0.30 +
    bowlers['match_score'] * 0.20
) * 100
```

```python
#Auction Price Prediction
bowlers['estimated_price_cr'] = (
    0.2 +
    (bowlers['bowler_score'] / 100) * 24.8
)

bowlers['estimated_price_cr'] = (
    bowlers['estimated_price_cr']
    .round(2)
)
```

```python
#Top Valuable Bowlers
top_bowlers = bowlers.sort_values(
    'estimated_price_cr',
    ascending=False
)[[
    'bowler',
    'wickets',
    'economy',
    'matches',
    'bowler_score',
    'estimated_price_cr'
]]

top_bowlers.head(15)
```

```python
#CREATE ALL-ROUNDER DATASET
all_rounders = player_stats[
    (player_stats['total_runs'] >= 1000) &
    (player_stats['wickets'] >= 30)
].copy()

all_rounders.head()
```

```python
# MERGE ECONOMY DATA FROM BOWLER DATASET
all_rounders = all_rounders.merge(
    bowler_stats[['bowler','economy']],
    left_on='player',
    right_on='bowler',
    how='left'
)

all_rounders.drop(columns=['bowler'], inplace=True)

all_rounders.head()
```

```python
#NORMALIZE FEATURES
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

all_rounders[['runs_score',
              'sr_score',
              'wicket_score',
              'match_score']] = scaler.fit_transform(

    all_rounders[['total_runs',
                  'strike_rate',
                  'wickets',
                  'matches']]
)

all_rounders.head()
```

```python
# ECONOMY SCORE
# Lower Economy = Better

all_rounders['economy_score'] = (
    all_rounders['economy'].max()
    - all_rounders['economy']
)

all_rounders[['economy_score']] = scaler.fit_transform(
    all_rounders[['economy_score']]
)
```

```python
# ALL-ROUNDER SCORE CALCULATION
# Weightage:
# Runs      = 30%
# StrikeRate= 15%
# Wickets   = 30%
# Economy   = 15%
# Matches   = 10%

all_rounders['all_rounder_score'] = (

    all_rounders['runs_score'] * 0.30 +

    all_rounders['sr_score'] * 0.15 +

    all_rounders['wicket_score'] * 0.30 +

    all_rounders['economy_score'] * 0.15 +

    all_rounders['match_score'] * 0.10

) * 100
```

```python
# AUCTION PRICE ESTIMATION
# Minimum Price = ₹20 Lakh
# Maximum Price = ₹25 Crore

all_rounders['estimated_price_cr'] = (

    0.20 +

    (all_rounders['all_rounder_score'] / 100) * 24.80

).round(2)
```

```python
# TOP VALUABLE ALL-ROUNDERS
top_all_rounders = all_rounders.sort_values(
    'estimated_price_cr',
    ascending=False
)[[
    'player',
    'total_runs',
    'strike_rate',
    'wickets',
    'economy',
    'matches',
    'all_rounder_score',
    'estimated_price_cr'
]]

top_all_rounders.head(20)
```

```python
# ============================================================
# IPL ALL-ROUNDER VALUATION MODEL
# ============================================================

# BUSINESS PROBLEM:
#
# All-rounders are among the most valuable players in IPL because
# they contribute in both batting and bowling departments.
#
# However, evaluating an all-rounder's true value is difficult
# because franchises must assess both batting and bowling
# performance simultaneously.
#
# A player may score many runs but contribute little with the ball,
# while another player may take wickets but have limited batting impact.
#
# To address this challenge, a Data-Driven All-Rounder Valuation
# Model was developed using historical IPL batting and bowling data.

# ============================================================
# OBJECTIVE
# ============================================================
#
# To estimate the fair auction value of an all-rounder based on:
#
# 1. Total Runs Scored
# 2. Strike Rate
# 3. Total Wickets Taken
# 4. Economy Rate
# 5. Matches Played (Experience)
#
# The model helps franchises identify high-impact players who
# contribute in multiple aspects of the game.

# ============================================================
# FEATURES USED
# ============================================================
#
# 1. TOTAL RUNS (Weight: 30%)
#    - Measures batting contribution and consistency.
#    - Higher runs indicate greater impact with the bat.
#
# 2. STRIKE RATE (Weight: 15%)
#    - Measures scoring speed.
#    - Higher strike rate indicates aggressive batting.
#
# 3. TOTAL WICKETS (Weight: 30%)
#    - Measures bowling contribution.
#    - Higher wickets indicate stronger bowling impact.
#
# 4. ECONOMY RATE (Weight: 15%)
#    - Measures bowling efficiency.
#    - Lower economy indicates better control over runs.
#
# 5. MATCHES PLAYED (Weight: 10%)
#    - Represents player experience and reliability.
#    - Experienced players are more effective under pressure.

# ============================================================
# ECONOMY RATE CALCULATION
# ============================================================
#
# Economy Rate is used to measure bowling efficiency.
#
# Formula:
#
# Economy Rate = Runs Conceded / Overs Bowled
#
# Step 1:
# Convert total balls bowled into overs.
#
# Overs Bowled = Balls Bowled / 6
#
# Step 2:
# Calculate economy rate.
#
# Economy Rate = Runs Conceded / Overs Bowled
#
# Example:
#
# Runs Conceded = 300
# Balls Bowled = 240
#
# Overs Bowled = 240 / 6 = 40
#
# Economy Rate = 300 / 40 = 7.5
#
# Interpretation:
#
# Economy 6-7   = Excellent
# Economy 7-8   = Good
# Economy 8-9   = Average
# Economy > 9   = Poor
#
# Since lower economy is better, reverse scoring was applied
# so that players with lower economy receive higher scores.

# ============================================================
# PLAYER SCORE CALCULATION
# ============================================================
#
# All performance metrics were normalized using MinMax Scaling
# to bring them onto the same scale.
#
# All-Rounder Score Formula:
#
# All-Rounder Score =
#
# (Runs Score × 30%)
# + (Strike Rate Score × 15%)
# + (Wicket Score × 30%)
# + (Economy Score × 15%)
# + (Match Experience Score × 10%)
#
# Final score is calculated on a scale of 0–100.

# ============================================================
# AUCTION PRICE ESTIMATION
# ============================================================
#
# IPL auctions have a minimum base price for players.
#
# Assumptions:
#
# Minimum Auction Price = ₹20 Lakh (0.20 Cr)
# Maximum Auction Price = ₹25 Crore
#
# Estimated Auction Price Formula:
#
# Estimated Price (Cr) =
#
# 0.20 + (All-Rounder Score / 100) × 24.80
#
# This ensures:
#
# - No player value falls below the IPL base price.
# - Elite all-rounders receive realistic auction valuations.
# - Better overall performance results in higher auction value.

# ============================================================
# BUSINESS INSIGHTS
# ============================================================
#
# Players with:
#
# - Strong batting performance
# - High strike rates
# - Consistent wicket-taking ability
# - Low economy rates
# - Extensive IPL experience
#
# receive higher valuation scores and estimated auction prices.
#
# Examples:
#
# Ravindra Jadeja
# Hardik Pandya
# Andre Russell
# Sunil Narine
# Shane Watson
#
# These players are expected to receive premium auction
# valuations because they contribute significantly in both
# batting and bowling departments.
#
# The model rewards complete cricketers who provide value
# across multiple phases of the game.

# ============================================================
# BUSINESS VALUE
# ============================================================
#
# This model helps IPL franchises:
#
# ✓ Identify high-impact all-rounders
# ✓ Optimize auction spending
# ✓ Reduce investment risk
# ✓ Improve team balance
# ✓ Make data-driven decisions
#
# Since all-rounders contribute in multiple roles,
# they often provide a higher return on investment
# compared to specialist players.

# ============================================================
# CONCLUSION
# ============================================================
#
# This All-Rounder Valuation Model converts historical
# batting and bowling performance into an estimated
# auction value.
#
# By combining batting ability, bowling effectiveness,
# and experience, the model provides a comprehensive
# assessment of player value.
#
# Instead of relying solely on reputation or recent form,
# franchises can use objective performance metrics to
# identify the most valuable all-rounders in the IPL.
#
# ============================================================
```
